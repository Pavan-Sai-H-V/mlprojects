```python
from fuzzywuzzy import process

standard_colleges = ["Hindustan University/HITS", "V.S.B.ENGINEERING COLLEGE,KARUR", "Sri Sivasubramaniya Nadar","IFET COLLEGE OF ENGINEERING","SRI KRISHNA COLLEGE OF TECHNOLOGY","Manakula Vinayagar Institute of Technology","St. Joseph's institute of technology","SRMIST-SRM Institute of Science and Technology"]

# Function to match each entry to the closest name
def match_college(name):
    match, score = process.extractOne(name, standard_colleges)
    return match if score > 80 else name  # Adjust threshold as needed

df["College"] = df["College"].apply(match_college)

print(df)
```
