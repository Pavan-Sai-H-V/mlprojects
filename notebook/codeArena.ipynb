```python
import pandas as pd
```

```python
df =pd.read_csv("data\CodeArena-registrations_2025-03-18.csv")
```

```python
df.columns
```

```python
df['College'].value_counts()
```

```python
from fuzzywuzzy import process

standard_colleges = ["Hindustan University/HITS", "V.S.B.ENGINEERING COLLEGE,KARUR", "Sri Sivasubramaniya Nadar","IFET COLLEGE OF ENGINEERING","SRI KRISHNA COLLEGE OF TECHNOLOGY","Manakula Vinayagar Institute of Technology","St. Joseph's institute of technology","SRMIST-SRM Institute of Science and Technology","Easwari Engineering college"]

# Function to match each entry to the closest name
def match_college(name):
    match, score = process.extractOne(name, standard_colleges)
    return match if score > 80 else name  # Adjust threshold as needed

df["College"] = df["College"].apply(match_college)

print(df)
```

```python
df.College.value_counts()
```

```python
filtered_df = df[~df["College"].str.contains("Hindustan University/HITS", case=False, na=False)]
filtered_df = filtered_df.reset_index(drop=True)
team_lead_phones = filtered_df[["College","Team Lead","Phone"]].reset_index(drop=True)
print(team_lead_phones)
```

```python
filtered_df.to_csv("External_colleges_CodeArena.csv", index=False)
```
