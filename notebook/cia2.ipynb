20) - Create a scatterplot using Python to show the relationship between two
quantitative variables     

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    "Hours Studied": [2, 5, 8],
    "Exam Score": [55, 70, 90]
}
df = pd.DataFrame(data)

# Scatterplot
sns.scatterplot(data=df, x="Hours Studied", y="Exam Score", color="blue")
plt.title("Scatterplot of Hours Studied vs Exam Score")
plt.show()
```

19) - Implement a correlation heatmap (Correlogram) for a dataset and analyze the relationships.

```python
# Data
data = {
    "Experience (Years)": [1, 3, 5, 7],
    "Hours Worked/Week": [35, 40, 45, 50],
    "Salary ($)": [40000, 55000, 70000, 85000],
    "Projects Handled": [2, 3, 4, 5]
}
df = pd.DataFrame(data)

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

12) Discuss the advantages and potential challenges of using color scales. Use a 
dataset of temperature variations across seasons.


```python
# Time-Series Data
data = {
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03"],
    "Temperature": [25, 22, 30],
    "Rainfall": [12, 5, 0],
    "Wind Speed": [15, 10, 20]
}
df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])

# Line Plot
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x="Date", y="Temperature", label="Temperature (C)", marker="o")
sns.lineplot(data=df, x="Date", y="Rainfall", label="Rainfall (mm)", marker="s")
sns.lineplot(data=df, x="Date", y="Wind Speed", label="Wind Speed (km/h)", marker="^")
plt.legend()
plt.title("Time-Series Data")
plt.show()
```

15) Explain and implement time-series decomposition to analyse trends in stock market data.

```python
import statsmodels.api as sm

# Stock Market Data
data = {
    "Date": ["2024-03-01", "2024-03-02", "2024-03-03"],
    "Closing Price ($)": [150, 152, 148]
}
df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)

# Decomposition
decomposed = sm.tsa.seasonal_decompose(df["Closing Price ($)"], model='additive', period=1)
decomposed.plot()
plt.show()
```

13) Develop a grid chart to compare multiple graphs using Matplotlib.

```python
# Sales Data
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Product A Sales": [500, 520, 550, 580],
    "Product B Sales": [400, 410, 420, 440],
    "Product C Sales": [300, 320, 350, 370],
})

# FacetGrid
df_melted = df.melt(id_vars=["Month"], var_name="Product", value_name="Sales")
g = sns.FacetGrid(df_melted, col="Product", col_wrap=2, height=4, sharey=False)
g.map_dataframe(sns.lineplot, x="Month", y="Sales", marker="o")
g.set_titles("{col_name}")
plt.show()
```
